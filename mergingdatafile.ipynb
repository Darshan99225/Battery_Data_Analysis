```python
import pandas as pd
```

```python
metadata_df = pd.read_csv("metadata.csv")
metadata_df
```

```python
file_name = metadata_df.loc[metadata_df['type']=='discharge',"filename"].values[0]
file_data = pd.read_csv("data/"+file_name)
file_data = pd.DataFrame(columns=file_data.columns.values)
file_data = file_data.reindex(columns = file_data.columns.tolist() + ['battery_id'])
file_data
```

```python
file_name = metadata_df.loc[metadata_df['type']=='discharge',"filename"].values[0]
file_data = pd.read_csv("data/"+file_name)
file_data = file_data.reindex(columns = file_data.columns.tolist() + ['battery_id'])
file_data.to_csv("discharge.csv",mode='a')
for id, rows in metadata_df.iterrows():
    if(rows["type"]=="discharge"):
        file_data = pd.read_csv("data/"+rows["filename"])
        file_data["battery_id"] = rows["battery_id"]
        file_data = file_data.rename(columns=file_data.iloc[0]).drop(file_data.index[0])
        file_data.to_csv("discharge.csv",mode='a')
```
